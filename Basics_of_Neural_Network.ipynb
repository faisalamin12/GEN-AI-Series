# Basics of Working of Neural Networks

A Neural Network is a model that learns patterns from data and makes predictions.

Inspired by how the human brain works.

### Basic Structure

A neural network has 3 main parts:

Input Layer → Takes data
Hidden Layer(s) → Processes data
Output Layer → Gives result

~~~
Input → Hidden → Output
~~~


## How a Neuron Works

1. **Multiply inputs with weights**
~~~
z=w1​x1​+w2​x2​+...+b
~~~

𝑥= inputs

𝑤= weights

𝑏= bias

2. Apply Activation Function


~~~
𝑎 = 𝑓(𝑧)
~~~

This helps the model learn complex patterns.

Common functions:

* ReLU
* Sigmoid

4. Forward Propagation (Prediction Step)

Data flows from input to output:
~~~
Input → Calculation → Activation → Output
~~~

This is how prediction is made.


3. Loss Function (Error)

After prediction, we compare:

~~~
Error = Actual − Predicted
~~~

Loss tells how wrong the model is.


4. Backpropagation (Learning Step)

The model learns by:

* Sending error backward
* Adjusting weights

7. Updating Weights

Weights are updated using:

𝑤 = 𝑤 − 𝛼 × gradient

α = learning rate

# Practical Implementation (Simple Neural Network)

```python
import numpy as np
```

```python
# Inputs
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

# Outputs
y = np.array([[0],
              [1],
              [1],
              [0]])
```

```python
np.random.seed(42)

weights = np.random.rand(2,1)
bias = np.random.rand(1)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
learning_rate = 0.1

for epoch in range(10000):

    # Forward Propagation
    z = np.dot(X, weights) + bias
    y_pred = sigmoid(z)

    # Loss (Error)
    error = y - y_pred

    # Backpropagation (Gradient)
    d_weights = np.dot(X.T, error * (y_pred * (1 - y_pred)))
    d_bias = np.sum(error * (y_pred * (1 - y_pred)))

    # Update Weights
    weights += learning_rate * d_weights
    bias += learning_rate * d_bias
```
