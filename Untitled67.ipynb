#Activation Function

In a neural network, after we calculate the weighted sum of inputs, the result is not sent directly to the next layer. Instead, it is passed through something called an activation function.

An activation function is a mathematical function that decides whether a neuron should be activated or not, and how much signal it should pass forward.

In simple terms, it acts like a decision-maker for each neuron.

### *Why do we need Activation Functions?*

If we do not use activation functions, the neural network will behave like a simple linear model, no matter how many layers we add. This means it will not be able to learn complex patterns.

### *Working of Activation Function*

An activation function works by taking the total input a neuron receives (after multiplying inputs with weights and adding bias) and then deciding how much of that value should be passed to the next layer. It transforms the input into a new output, often by scaling it or limiting it within a range, so the network can learn patterns properly. In simple words, it acts like a filter or decision-maker that controls the strength of the signal moving forward in the neural network.

### *How Activation Function decides the output*

An activation function decides the output by applying a mathematical rule to the input value (weighted sum + bias). It checks the input and then transforms it into a specific range or form. For example, some activation functions only allow positive values, some convert the output between 0 and 1, and some scale it differently. Based on this rule, the function determines how strong or weak the signal should be, and then passes that adjusted output to the next layer.

###**Example**

Suppose a neuron receives inputs like study hours and attendance of a student. First, each input is multiplied by its importance (weights), and then all values are added together along with a bias to get a final number (for example, 3.5). Now, this value is passed to the **activation function**, which decides what to do with it maybe it converts it into a value between 0 and 1 (like 0.97) to represent confidence. This output shows how strongly the neuron is activated and is then sent to the next layer. In simple words, the activation function takes the calculated value, applies its rule, and decides how much signal should move forward.


## NAMES OF ACTIVATION FUNCTIONS WITH ITS WORKING

### **Common Activation Functions (List)**

* Sigmoid
* Tanh (Hyperbolic Tangent)
* ReLU (Rectified Linear Unit)
* Leaky ReLU
* Softmax


### **1. Sigmoid Function**

The **Sigmoid** function converts any input value into a range between **0 and 1**. It is often used when we want probability-like outputs. For example, if a neuron calculates a value like 3, the sigmoid function may convert it to 0.95, meaning a high probability. If the value is -3, it may become 0.05, meaning low probability. Imagine predicting whether a student will pass or fail: if the output is close to 1, it means “pass”; if close to 0, it means “fail”. So, sigmoid helps in making decisions based on probability.


### **2. Tanh (Hyperbolic Tangent)**

The **Tanh** function converts input values into a range between **-1 and 1**. It is similar to sigmoid but better because it handles both positive and negative values. For example, if the input is 2, tanh might give 0.96, and if the input is -2, it might give -0.96. Think of it like measuring mood: positive values mean positive response, negative values mean negative response, and values near 0 mean neutral. This helps the network learn better balanced patterns.


### **3. ReLU (Rectified Linear Unit)**

The **ReLU** function is very simple: it outputs the same value if it is positive, but if the value is negative, it returns 0. For example, if the input is 5, output is 5; if input is -3, output is 0. Imagine it like a filter that blocks all negative signals and only allows positive ones. This makes computation faster and helps deep networks learn efficiently.


### **4. Leaky ReLU**

The **Leaky ReLU** is an improved version of ReLU. Instead of making all negative values 0, it allows a small portion of them to pass. For example, if the input is -4, instead of 0, it might give -0.04. This prevents neurons from completely “dying” (stopping learning). Think of it like a slightly open door for negative values instead of completely shutting them off.


### **5. Softmax Function**

The **Softmax** function is used when there are multiple classes. It converts outputs into probabilities that add up to 1. For example, if we are classifying an image as Cat, Dog, or Bird, softmax might give outputs like: Cat = 0.7, Dog = 0.2, Bird = 0.1. This means the model is most confident that the image is a Cat. It helps in choosing one class out of many by comparing probabilities.

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = 3
print(sigmoid(x))
```

The sigmoid function takes any number and converts it into a value between 0 and 1. In this example, input 3 will give an output close to 0.95. This means the model is highly confident about something (like predicting “Yes” or “True”). It is mainly used when we need probability outputs, such as in binary classification.

```python
import numpy as np

def tanh(x):
    return np.tanh(x)

x = 2
print(tanh(x))
```

The tanh function converts values into a range between -1 and 1. Here, input 2 gives output close to 0.96. Unlike sigmoid, it handles both positive and negative values, making it better for balanced data. It helps the model understand both positive and negative signals.

```python
def leaky_relu(x):
    return x if x > 0 else 0.01 * x

x = -4
print(leaky_relu(x))
```

The Leaky ReLU is similar to ReLU but allows a small negative value instead of zero. Here, input -4 becomes -0.04. This helps prevent neurons from completely stopping learning. It is useful when ReLU causes problems like “dead neurons”.

```python
import numpy as np

def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x)

x = np.array([2.0, 1.0, 0.1])
print(softmax(x))
```

The softmax function converts multiple values into probabilities that sum to 1. In this example, it will output something like [0.7, 0.2, 0.1]. This means the first class has the highest probability. It is mainly used in multi-class classification problems.
