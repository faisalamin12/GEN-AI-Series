# Difference 1: Feature Engineering (ML vs DL)

### Machine Learning (Feature Engineering Required)

In Machine Learning, models cannot automatically understand complex relationships in raw data. Therefore, we often perform feature engineering, where we manually create new features to help the model learn better patterns.

````python
import numpy as np

# Create new features manually
X_train_fe = np.c_[X_train, X_train[:,0]**2, X_train[:,1]**2]
X_test_fe = np.c_[X_test, X_test[:,0]**2, X_test[:,1]**2]
````

Here, we manually added polynomial features (squared terms). These new features help transform the data into a space where a linear model can perform better.


*   Logistic Regression cannot handle non-linear patterns directly
*   We artificially introduce non-linearity through feature engineering



````python
from sklearn.linear_model import LogisticRegression
ml_model_fe = LogisticRegression()
ml_model_fe.fit(X_train_fe, y_train)

ml_accuracy_fe = ml_model_fe.score(X_test_fe, y_test)
print("ML Accuracy after Feature Engineering:", ml_accuracy_fe)
````

After adding new features, you will notice an improvement in accuracy (often 85%–90%).

This shows that:

* ML performance heavily depends on feature quality
* Without feature engineering → lower accuracy
* With feature engineering → better results

# Difference 3: GPU Usage (ML vs DL)

In traditional Machine Learning, most algorithms are designed to run efficiently on CPUs. These models involve relatively simpler mathematical operations and smaller datasets, so they do not require massive parallel computation.

Algorithms like Logistic Regression, Decision Trees, and SVMs are typically optimized for CPU execution, and in most real-world cases, using a GPU does not significantly improve performance.


**Step: Train ML Model (CPU-based)**

````python
from sklearn.linear_model import LogisticRegression

ml_model = LogisticRegression()
ml_model.fit(X_train, y_train)
````

Here, the model is trained using CPU by default. Libraries like scikit-learn do not natively utilize GPUs for most algorithms.

Even if you run this code on a machine with a GPU:

* It will still use CPU
* Training time will not change significantly

This is because:

* ML models are not heavily parallel
* Computation is relatively lightweight

### Deep Learning (Heavy GPU Usage)

Deep Learning models, on the other hand, involve millions of parameters and require massive matrix computations. These operations can be parallelized, making GPUs extremely effective.

A GPU can process thousands of operations simultaneously, which significantly speeds up training.


**Step: Check GPU Availability**

````python
import tensorflow as tf

print("GPU Available:", tf.config.list_physical_devices('GPU'))
````

This checks whether TensorFlow can access a GPU. If a GPU is available, TensorFlow will automatically use it for training.

### Train DL Model (GPU Accelerated)

````python
with tf.device('/GPU:0'):
    history = dl_model.fit(
        X_train, y_train,
        epochs=50,
        batch_size=32,
        validation_split=0.1
    )
````

Here, training is explicitly assigned to the GPU.

During training:

* Forward propagation → runs on GPU
* Backpropagation → runs on GPU
* Matrix multiplications → massively parallel

This results in:

* Faster training
* Ability to handle large datasets
* Efficient deep architectures

| Concept             | Machine Learning | Deep Learning   |
| ------------------- | ---------------- | --------------- |
| Hardware            | CPU              | GPU (preferred) |
| Computation         | Light            | Heavy           |
| Parallelism         | Low              | High            |
| Speed Gain from GPU | Minimal          | Huge            |
